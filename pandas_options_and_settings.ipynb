# Pandas options and settings
- change number of columns and rows displayed in a DataFrame
- change number of digits after decimal point in a float that are displayed
  - but complexity is preserved

```python
# libraries needed
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# 1000 rows x 50 column filled with integers between 0 (inclusive) and 100 (exclusive)
random_data = np.random.randint(
    low = 0, 
    high = 100, 
    size = [1000, 50]
  )

random_data
```

```python
# convert random_data to a DataFrame
df = pd.DataFrame(random_data)

df.head()
```

## change number of rows and columns displayed in a DataFrame

```python
# number of rows displayed is 60 by default; but doesn't play nice with google colab
pd.options.display.max_rows

df
```

```python
# change default
pd.options.display.max_rows = 6

df
```

```python
# same thing with columns
pd.options.display.max_columns
```

```python
df
```

```python
# see all columns
pd.options.display.max_columns = 50
df
```

## change number of rows or columns displayed in a DataFrame, using methods, functionally same as above
- pd.get_option( ) to see current setting
  - parameter is 'max_columns' or 'max_rows'
- pd.set_option( ) to change setting
  - parameter is 'max_columns' or 'max_rows' then a number
- pd.reset_option( ) to restore defaults
  - parameter is 'max_columns' or 'max_rows'
- pd.describe_option( ) to get some documentation
  - parameter is 'max_columns' or 'max_rows'


```python
# maximum number of rows displayed
pd.get_option('max_rows')
```

```python
# maximum number of columns displayed
pd.get_option('max_columns')
```

```python
# change number of rows
pd.set_option('max_rows', 10)

# now we see that many rows displayed
df
```

```python
# change number of columns
pd.set_option('max_columns', 12)

# now we see that many columns
df
```

```python
# reset defaults
pd.reset_option('max_rows')
pd.reset_option('max_columns')
```

```python
# view defaults
print(pd.get_option('max_rows'))
print(pd.get_option('max_columns'))
```

```python
pd.describe_option('max_rows')
```

```python
pd.describe_option('max_columns')
```

## precision
- changes number of digits shown for floats
- doesn't modify, just appearance

```python
# fake data
fake_data = pd.DataFrame(
    data = np.random.randn(5, 5)         # 5 x 5 array; randomly drawn from normal distribution of mean 0 and variance of 1
)

fake_data
```

```python
pd.get_option('precision')   # 6 numbers after the decimal
```

```python
# view fake_data, but change precisoin to just 2 digits, i.e., everything is rounded
pd.set_option('precision', 2)

fake_data
```

```python
fake_data.iloc[0, 0]

# complexity is preserved
```

```python
# restore default
pd.reset_option('precision')

fake_data
```
