# Pandas groupby

```python
# libraries needed
import numpy as np
import pandas as pd
```

```python
# load data
fortune_1k = pd.read_csv(
    'https://raw.githubusercontent.com/jack-cao-623/python_learning/main/pandas/fortune1000.csv'
)

fortune_1k.head()
```

```python
# meta data
fortune_1k.info()
```

```python
# make 'Rank' the index
fortune_1k = fortune_1k.set_index('Rank')
```

```python
fortune_1k.head()
```

```python
# total profit by sector
(
    fortune_1k
      .groupby('Sector')
      ['Profits']
      .sum()
      .to_frame()
      .reset_index()
      .sort_values(by = 'Profits', ascending = False)
)
```
