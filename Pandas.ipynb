```python
# libraries needed
import numpy as np
import pandas as pd
```

```python
# load data into python
wine = pd.read_csv('winemag-data-130k-v2.csv', index_col = 0)

# first 5 rows
wine.head()
```

```python
# number of rows and columns
wine.shape
  # 130k rows, 13 columns
```

```python
# column names
wine.columns
```

```python
# number of rows for each country
wine['country'].value_counts()
```

```python
# rows where country is U.S. and points is >= 90
wine[(wine['country'] == 'US') & (wine['points'] >= 90)]
```

```python
# using .iloc[]

# first 12 rows and first 2 columns
wine.iloc[:12, :2]

# don't use brackets [] or parentheses () with .iloc[]
```

```python
# with .loc[]

# first 12 rows and first 2 columns, country and description

wine.loc[:12, ['country', 'description']]
```

```python
# first 10 rows using .iloc[]
wine.iloc[:10, :]   # all columns, be explicit
```

```python
# first 10 rows using .loc[]
wine.loc[:9, :]
```

```python
# filter to rows where region_1 is Willamette Valley using .query
(
    wine
      .query("region_1 == 'Willamette Valley'")   # filter to rows where region_1 equals Willamette Valley
      ['country']                                 # pull out country as a series
      .unique()                                   # unique values in country series
)

#wine.query("region_1 == 'Willamette Valley'")
```

```python
# repeat above without using .query()
(
    wine
      [(wine['region_1'] == 'Willamette Valley') &
       (wine['points'] >= 90)]
      #['country']
      #.unique()
)
```

```python
# average score per country
(
    wine
      .groupby('country')
      [['points']]
      .mean()
      #.index
      #.shape
      #.sort_values('points', ascending = False)
)
```

```python
# how many of each country
(
    wine
      .groupby('country')
      ['country']
      .count()
      .sort_values(ascending = False)
)
```

```python
# for each point, stats on price
(
    wine
      .groupby('points')
      ['price']
      .agg(['min', 'median', 'mean', 'max'])
      .reset_index()
      .sort_values(by = 'points', ascending = False)
)
```
