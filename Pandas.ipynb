```python
# libraries needed
import numpy as np
import pandas as pd
```

```python
# load data into python
wine = pd.read_csv('winemag-data-130k-v2.csv', index_col = 0)

# first 5 rows
wine.head()
```

```python
# number of rows and columns
wine.shape
  # 130k rows, 13 columns
```

```python
# column names
wine.columns
```

```python
# number of rows for each country
wine['country'].value_counts()
```

```python
# rows where country is U.S. and points is >= 90
(
    wine
    # filter
    [(wine['country'] == 'US')     # only United States
      & (wine['points'] >= 90)     # scored at least 90 points
    ]
    .shape
)

#wine[(wine['country'] == 'US') & (wine['points'] >= 90)]
```

```python
# using .iloc[]

# first 12 rows and first 2 columns
wine.iloc[:12, :2]

# don't use brackets [] or parentheses () with .iloc[]
```

```python
# with .loc[]

# first 13 rows and first 2 columns, country and description

wine.loc[:12, ['country', 'description']]
```

```python
# first 10 rows using .iloc[]
wine.iloc[:10, :]   # all columns, be explicit
```

```python
# first 10 rows using .loc[]
wine.loc[:9, :]
```

```python
# filter to rows where region_1 is Willamette Valley using .query
(
    wine
      .query("region_1 == 'Willamette Valley'")   # filter to rows where region_1 equals Willamette Valley
      ['country']                                 # pull out country as a series
      .unique()                                   # unique values in country series
)

#wine.query("region_1 == 'Willamette Valley'")
```

```python
# repeat above without using .query()
(
    wine
      [(wine['region_1'] == 'Willamette Valley') &
       (wine['points'] >= 90)]
      #['country']
      #.unique()
)
```

```python
# average score per country
(
    wine
      .groupby('country')
      [['points']]
      .mean()
      #.index
      #.shape
      #.sort_values('points', ascending = False)
)
```

```python
# how many of each country
(
    wine
      .groupby('country')
      ['country']
      .count()
      .sort_values(ascending = False)
)
```

```python
# for each point, stats on price
(
    wine
      .groupby('points')
      ['price']
      .agg(['min', 'median', 'mean', 'max', 'count'])
      .reset_index()
      .sort_values(by = 'points', ascending = False)
)
```

```python
# for each winery, title of first wine
(
    wine
      .groupby('winery')
      .apply(lambda row: row['title'].iloc[0])
)
```

```python
# for each country and province,what's the best wine
(
    wine
      .groupby(['country', 'province'])
      .apply(lambda x: x.loc[x['points'].idxmax()])    # why loc instead of iloc?
)
```

```python
# number of provinces per country
(
    wine
      .groupby('country')
      ['region_2']
      .agg(['size', 'count'])
      .sort_index()
      #.sort_values(ascending = False)
)
```

```python
# for each taster, what was their best and worst score given
(
    wine
      .groupby('taster_name')
      ['points']
      .agg(['min', 'max'])
      .sort_index()
)
```

```python
wine.columns
```

```python
# tasters that gave worst (80) and best (100) score
tasters_all = (
    wine
      .groupby('taster_name')
      ['points']
      .agg(['min', 'max'])
      .rename(columns = {'min': 'min_points', 'max': 'max_points'})
)
```

```python
# these four tasters gave min and max points
tasters_all[(tasters_all['min_points'] == 80) & (tasters_all['max_points'] == 100)]
```

```python
# string above together into one step
(
    wine
      .groupby('taster_name')                                         # one row per taster; this is the index
      ['points']                                                      # pull out points column
      .agg(['min', 'max'])                                            # for each taster, min and max points given
      .rename(columns = {'min': 'min_points', 'max': 'max_points'})   # rename columns
      .reset_index()                                                  # changes the index to a column
      .query("min_points == 80 & max_points == 100")                  # filters to tasters that gave lowest and highest score
)
```

```python
(
    wine
      .groupby('taster_name')                                         # one row per taster; this is the index
      ['points']                                                      # pull out points column
      .agg(['min', 'max'])                                            # for each taster, min and max points given
      .rename(columns = {'min': 'min_points', 'max': 'max_points'})   # rename columns
      .reset_index()                                                  # changes the index to a column
      #.query("min_points == 80 & max_points == 100")                  # filters to tasters that gave lowest and highest score
      .query("(max_points - min_points) != 20")
)
```
